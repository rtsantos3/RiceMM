```python
import cobra
import libsbml
import escher
from cobra import Model, Reaction, Metabolite
```

```python
ios2164_orig = cobra.io.read_sbml_model("../model/iOS2164_rewritten.xml")
ios2164_m = cobra.io.read_sbml_model("../model/iOS2164_rewritten.xml")
ios2164_bs = cobra.io.read_sbml_model("../model/iOS2164_rewritten.xml")
```

```python
#initialize empty cobra model for 2-cell
ios2164_2cell = Model()
```

```python
for item in ios2164_m.metabolites:
    item.id = str(item.id) + '0'
    item.name = str(item.name) + '_' + 'M'
    item.compartment = str(item.compartment) + '0'

    ios2164_m.compartments = {'c0': 'M Cell Cytoplasm',
 's0': 'M Cell Plastid',
 'r0': 'M Cell Endoplasmic_reticulum',
 'm0': 'M Cell Mitochondrion',
 'e0': 'M Cell Extracellular',
 'u0': 'M Cell Thylakoid',
 'x0': 'M Cell Peroxisome',
 'v0': 'M Cell Vacuole'}
```

```python
for item in ios2164_bs.metabolites:
    item.id = str(item.id) + '1'
    item.name = str(item.name) + '_' + 'BS'
    item.compartment = str(item.compartment) + '1'

    ios2164_bs.compartments = {'c1': 'BS Cell Cytoplasm',
 's1': 'BS Cell Plastid',
 'r1': 'BS Cell Endoplasmic_reticulum',
 'm1': 'BS Cell Mitochondrion',
 'e1': 'BS Cell Extracellular',
 'u1': 'BS Cell Thylakoid',
 'x1': 'BS Cell Peroxisome',
 'v1': 'BS Cell Vacuole'}
    
```

```python
ios2164_2cell = ios2164_m.merge(ios2164_bs, objective = "sum", inplace =  1)
```

From Blatke & Brautigam (2019):
"The mesophyll and bundle sheath networks are connected by a range of cytosolic transport metabolites including amino acids, sugars (glucose, fructose, sucrose, trehalose, ribose), single phosphorylated sugar (glucose-6-phosphate, glucose-1-phosphate, fructose-6-phosphate, sucrose-6-phosphate), mono-/di-/tri-carboxylic acids (phosphoenolpyruvate, pyruvate, citrate, cis-aconitate, isocitrate, α-ketoglutarate, succinate, fumarate, malate), glyceric acids (2-Phosphoglycerate, 3-Phosphoglycerate), glycolate, glycerate, glyceraldehyde-3-phosphate, di-hydroxyacetone-phosphate and CO2. Nucleotides, NAD/NADH, NADP/NADPH, pyrophosphate, inorganic phosphate are not considered as transport metabolites. Oxaloacetate has been excluded as transport metabolite since concentrations of oxaloacetate are very low in vivo and it is reasonably unstable in aqueous solutions."
