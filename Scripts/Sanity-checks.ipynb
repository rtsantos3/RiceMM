```python
DM_atp = cobra.Reaction('DM_atp')
DM_atp.add_metabolites({model.metabolites.atp_c:-1})


with model:
    #Turn off all boundary reactions (Closed system)
    for item in model.boundary:
        item.bounds = (0,0)
    model.add_reaction(DM_atp)
    #Check forward (from cell to outside)
    model.reactions.DM_atp.bounds = (0, 1000)
    DM_atp_fw = model.optimize()
    #Check backward (From outside to cell)
    
    model.reactions.DM_atp.bounds = (-1000, 0)
    DM_atp_bw = model.optimize()
    
    print('ATP fw: ', DM_atp_fw.status, 'ATP bw: ', DM_atp_bw.status)
```

ATP drain/input tests show that solution is infeasible if applied to a closed system, meaning that it doesn't produce any biomass if ATP is supplied/drained into the system.
